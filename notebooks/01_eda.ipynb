# Premier League EDA

Quick exploratory pass before/after feature engineering.

```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(ROOT))

from src.load_data import load_matches

matches = load_matches()
matches.head()
```

```python
print(f"Matches: {len(matches)}")
print(matches["FTR"].value_counts(normalize=True).round(3))
```

```python
matches["total_goals"] = matches["FTHG"] + matches["FTAG"]
by_season = matches.groupby("Season")["FTR"].value_counts(normalize=True).unstack()
by_season
```

```python
home_rate = (matches["FTR"] == "H").groupby(matches["Season"]).mean()
home_rate.plot(kind="bar", title="Home win rate by season", ylabel="rate")
plt.tight_layout()
plt.show()
```
